# Normalizing And Standardizing Data <a id="normalizing-and-standardizing-data"></a>

```python
bikesales.info()
```

> Date assigned as object, day assigned as integer, year assigned as integer, customer age assigned as integer, unit cost, profit, cost and revenue assigned as integer all need to be converted to a proper data type 

#### Standardizing data types <a id="standardizing-data-types"></a>

#####   Day standardizing <a id="day-standardizing"></a>

```python
bikesales['Day'] = bikesales['Day'].astype('str')
```

#####   Year standardizing <a id="year-standardizing"></a>

```python
bikesales['Year'] = bikesales['Year'].astype('str')
```

```python
bikesales['Year']
```

##### Date standardizing <a id="date-standardizing"></a>

```python
bikesales['Date'] = pd.to_datetime(bikesales['Date'], dayfirst=False)
```

##### Customer age standardizing <a id="customer-age-standardizing"></a>

```python
bikesales['Customer_Age'] = bikesales['Customer_Age'].astype('str')
```

#### Standardizing other columns <a id="standardizing-other-columns"></a>

```python
bikesales[['Cost', 'Unit_Cost', 'Unit_Price', 'Profit', 'Revenue']] = bikesales[['Cost', 'Unit_Cost', 'Unit_Price', 'Profit', 'Revenue']].astype(float)
```
